```python
import pandas as pd
import geopandas as gpd
```

##### Mapa de coropletas que muestre la cantidad de casos positivos de COVID en cantones de Costa Rica.

```python
cantonal = gpd.read_file("../Cantonal.shp")
cantonal.plot()
```

```python
# Carga de casos positivos
covid_cantonal_positivos = pd.read_csv("../05_30_22_CSV_POSITIVOS.csv", 
                                       sep=";", 
                                       encoding="iso-8859-1") # para leer tildes y otros caracteres

# Reducción de columnas
covid_cantonal_positivos = covid_cantonal_positivos[["cod_provin", "cod_canton", "provincia", "canton", "30/05/2022"]]

# Eliminación de fila con valores nulos
covid_cantonal_positivos = covid_cantonal_positivos.dropna(how='all')

# Eliminación de fila con canton=="Otros"
covid_cantonal_positivos = covid_cantonal_positivos[covid_cantonal_positivos["canton"] != "Otros"]

# Cambio de nombre de columnas
covid_cantonal_positivos = covid_cantonal_positivos.rename(columns={"30/05/2022": "positivos"})

covid_cantonal_positivos
```

### Unión (JOIN) de las dos capas

```python
cantones_covid = cantonal.merge(covid_cantonal_positivos, on="cod_canton")
```

```python
cantones_covid
```

```python
cantones_covid.plot(column="positivos", 
                    legend=True, 
                    cmap="YlOrBr", 
                    scheme="quantiles", 
                    figsize=(20, 20))
```

### Mapa de coropletas que muestre la cantidad de casos positivos de COVID en provincias de Costa Rica.

```python
provincial = gpd.read_file("../Provincias.shp")
provincial.plot()
```

```python
covid_provincial_positivos = covid_cantonal_positivos.groupby(["cod_provin"])["positivos"].sum()
covid_provincial_positivos
```

```python
provincias_covid = provincial.merge(covid_provincial_positivos, on="cod_provin")
```

```python
provincias_covid
```

```python
provincias_covid.plot(column="positivos", 
                    legend=True, 
                    cmap="YlOrBr", 
                    scheme="quantiles", 
                    figsize=(20, 20))
```

```python
murcielagos = pd.read_csv("../murcielagos.csv", sep="\t")
```

```python
murcielagos
```

```python
# Geodataframe creado a partir del dataframe
murcielagos = gpd.GeoDataFrame(murcielagos, 
                               geometry=gpd.points_from_xy(murcielagos.decimalLongitude, 
                                                           murcielagos.decimalLatitude),
                               crs="EPSG:4326")

# Despliegue de los datos (incluyendo geometrías)
murcielagos
```

```python
# Mapa
murcielagos.plot(figsize=(20, 20), color = "purple")
```

```python
# Join espacial de las capas de ASP y murciélagos
cantones_murcielagos = cantonal.sjoin(murcielagos)

cantones_murcielagos
```

```python
cantones_murcielagos.info()
```

```python
# Conteo de especies en cada ASP
conteo_cantones_murcielagos = cantones_murcielagos.groupby("canton").species.nunique()
conteo_cantones_murcielagos = conteo_cantones_murcielagos.reset_index() # para convertir la serie a dataframe
conteo_cantones_murcielagos
```

```python
cantones_murcielagos = cantonal.merge(conteo_cantones_murcielagos, on="canton")
```

```python
cantones_murcielagos
```

```python
cantones_murcielagos.plot(column="species", 
                    legend=True, 
                    cmap="YlOrBr", 
                    scheme="quantiles", 
                    figsize=(20, 20))
```
